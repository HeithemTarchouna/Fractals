```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# A subset of Mandelbrot but instead of C changing it's kept constant
# F(Z) = Z^2 + C

def julia(n_rows, n_columns, iterations, cx, cy):
    x_cor = np.linspace(-2, 2, n_rows)
    y_cor = np.linspace(-2, 2, n_columns)
    x_len = len(x_cor)
    y_len = len(y_cor)
    output = np.zeros((x_len,y_len))
    c = complex(cx, cy)
    for i in range(x_len):
        for j in range(y_len):
            z = complex(x_cor[i], y_cor[j])
            count = 0
            for k in range(iterations):
                z = (z * z) + c
                count = count + 1
                if (abs(z) > 4):
                    break
            output[i,j] = count
        #print(int((i/x_len)*100),"% completed")

    #print(output)
    print("Generating Fractal..")
    plt.imshow(output.T, cmap='hot')
    plt.axis("off")
    plt.show()
```

```python
julia(1500,1500,150,-0.42,0.6)
```
